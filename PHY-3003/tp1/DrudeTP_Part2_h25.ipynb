VIII. _(5 points)_ &mdash; Vérifiez numériquement et graphiquement que l'amortissement de la quantité de mouvement moyenne des électrons suit l'équation exponentielle dérivée analytiquement en classe, soit $\langle p(t)\rangle =\langle p(t_0)\rangle \,e^{-t/\tau}$, et comparez-y le comportement d'un seul électron.

```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("data_p_pt2.txt")
time = data[:,0]
tau = data[:,1]
p_single_particle = data[:,2]
p_avg = data[:,3]

p_t0 = np.sqrt(2*9.1E-31*(2/2)*1.4E-23*300)
t_theorique = np.linspace(time[0], time[-1], 1000)
tau_exp = np.mean(np.loadtxt("data_tau_pt2.txt"))
p_theorique = p_t0*np.exp(-t_theorique/tau_exp)
print(tau_exp, tau[0])

# On fit un tau expréimental:
from scipy.optimize import curve_fit
def expo(x, a,b,c):
    return np.exp(-(x)/a)*b+c
#def expo(x, a):
#    return np.exp(-x/a)
res = curve_fit(expo, time, p_avg/p_avg[0], p0=[1E-5,1,1])[0]
if False:
    def expo(x, a):
        return np.exp(-x/a)
    res = curve_fit(expo, time, p_avg/p_avg[0], p0=[1E-5,1,1])[0]
print(res)

fig = plt.figure(figsize=(15,5))
ax1 = plt.subplot(121)
ax1.plot(time, p_single_particle/p_single_particle[0], color="orange", linewidth=3, label="$ p$ particule")
ax1.plot(time, p_avg/p_avg[0], linestyle="-", color="blue", linewidth=2, label=r"$\langle p \rangle$")
time_theo = np.linspace(0,np.max(time),1000)
ax1.plot(time_theo, expo(time_theo, *res), linestyle="-", color="red", linewidth=2, label=r"$\exp[-t/\tau]$")
#plt.plot(time_theo, np.exp(-time_theo/tau_exp), linestyle="-", color="purple", linewidth=2, label="model forcé")
#plt.yscale("log")
#plt.xscale("log")
ax1.set_xlabel("Temps [s]")
ax1.set_ylabel("$p/p_0$")
ax1.legend()
ax2 = plt.subplot(122)
ax2.plot(time, (expo(time, *res)-(p_avg/p_avg[0]))/expo(time, *res)*100, ".", color="black")
ax2.hlines(0, time[0], time[-1], color="red", linestyles="--")
ax2.set_xlabel("Temps [s]")
ax2.set_ylabel("Résiduel [%]")
plt.show()
```

Nous observons que la quantité de mouvement moyenne $\langle p \rangle$, représentée par la courbe bleue suit effectivement une exponentielle $e^{-t/\tau}$, ajustée par la courbe rouge. Nous voyons que la quantité de mouvement d'une seule particule est plutôt aléatoire, mais diminue peu à peu. Cela est normal puisqu'à chaque collision d'une particule, sa quantité de mouvement est prise aléatoirement selon la distribution de Maxwell-Boltzmann correspondant à la température actuelle de la simulation.

Le panneau de droite montre les résidus de notre ajustement exponentiel à la quantité de mouvement moyenne des électrons en fonction du temps. Nous voyons que les résidus se situent toujours près de 0% d'erreur, atteingnant au maximum 2.5% d'écart avec nos données.

### Question statistique ###

X. _(5 points)_ &mdash; Pour quelques différents modules de champ électrique, présentez graphiquement l'évolution de la position moyenne des électrons en fonction du temps pour ses deux composantes parallèle et perpendiculaire au champ.

```python
import numpy as np
import matplotlib.pyplot as plt

modules = [0.1,0.01,0.001,0.2,0.05,1] # Modules que nous avons testés.
modules.sort() # Pour les avoir en ordre croissant.

t_list = []
x_list = []
y_list = []
for module in modules:
    filename = f"data_vec_xy_with_module_{module}_x.txt"
    data = np.loadtxt(filename)
    # Extraire les colonnes
    t = data[:, 0]  # Temps
    x = data[:, 1]  # Coordonnée x
    y = data[:, 2]  # Coordonnée y
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)

# Créer les figures
for i in range(len(modules)):
    fig, axs = plt.subplots(1, 1, figsize=(8, 3))
    # Tracer x(t)
    axs.plot(t_list[i], x_list[i], label="x(t)", color="b")
    # Tracer y(t)
    axs.plot(t_list[i], y_list[i], label="y(t)", color="r")
    axs.set_ylabel("Position [m]")
    axs.set_xlabel("Temps (s)")
    plt.title(f"Module du champ: $E={modules[i]}\cdot 10^{{12}} $ V/m")
    axs.legend()
    plt.show()
```

Nous remarquons que plus le module du champ électrique est élevé, plus la position moyenne en $x$ augmente rapidement et tend vers une position moyenne finale en $x$ élevée. Nous savons que cela est dû au champ électrique introduit puisque l'effet n'est pas observé en $y$. En effet, la position moyenne en $y$ reste stable autour de 0, ce qui est parfaitement normal. Nous observons aussi une petite oscillation quand le champ électrique est assez fort, qui correspond au rebond des électrons frappant le côté droit de la boîte. Lorsqu'ils vont trop vite, ils cognent le côté droit à une grande vitesse et rebondissent vers la gauche, ce qui fait redescendre la position moyenne en $x$, puis ainsi de suite, en amortissant de plus en plus le rebond.

Au contraire, lorsque le module est trop petit, la position $x$ moyenne ne semble pas être affectée et oscille autour de la position moyenne nulle, comme s'il n'y avait pas de champ électrique.
